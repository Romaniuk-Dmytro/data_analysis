```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture

mean1 = [3, 4]
cov1 = [[1, 0], [0, 1]]
x1 = np.random.multivariate_normal(mean1, cov1, 200)

mean2 = [5, 8]
cov2 = [[2, 0], [0, 2]]
x2 = np.random.multivariate_normal(mean2, cov2, 200)

x_combined = np.concatenate([x1, x2])

gm = GaussianMixture(n_components=2, random_state=0, covariance_type='full').fit(x_combined)

labels = gm.fit_predict(x_combined)
```

```python
ax = plt.gca()
ax.scatter(x_combined[:, 0], x_combined[:, 1], c=labels, s=40, cmap='viridis')
plt.show()
```

```python
x1 = np.random.default_rng().multivariate_normal(gm.means_[0], gm.covariances_[0], 200)
x2 = np.random.default_rng().multivariate_normal(gm.means_[1], gm.covariances_[1], 200)
df = np.concatenate([x1, x2])

ax = plt.gca()
ax.scatter(df[:, 0], df[:, 1], c=np.hstack([np.ones(200), np.zeros(200)]), s=40, cmap='viridis')
plt.show()
```

```python
print(gm.means_, '\n', mean1, mean2, '\n\n')
print(gm.covariances_, '\n', cov1, cov2)
```
