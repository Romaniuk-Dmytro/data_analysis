```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
np.random.seed(0)
X = np.random.rand(100, 1)
Y = 4.42 * X + 3.1 + np.random.rand(100, 1)
plt.scatter(X,Y,s=10)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# Gradient Descent Method
a = 0
b = 0

L = 0.1
epochs = 1000

n = float(len(X))

for i in range(epochs): 
    Y_pred = a*X + b
    D_a = (-2/n) * sum(X * (Y - Y_pred))
    D_b = (-2/n) * sum(Y - Y_pred)
    a = a - L * D_a
    b = b - L * D_b
    
print (a, b)


Y_pred = a*X + b

plt.scatter(X, Y,s=10) 
plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='red')
plt.show()
```
